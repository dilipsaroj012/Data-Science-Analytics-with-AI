# Data vesulization 

# 1:- Bar Graph

```python


# Assuming df is your DataFrame
top_10_sales = df.nlargest(10, 'Sales ($billion)')

plt.figure(figsize=(12, 6))
plt.bar(top_10_sales['Company'], top_10_sales['Sales ($billion)'])
plt.xlabel('Company')
plt.ylabel('Sales ($billion)')

plt.title('Top 10 Companies by Sales')
plt.xticks(rotation=45)
plt.show()
```

# 2:- Scatter plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales ($billion)'], df['Market Value ($billion)'])
plt.xlabel('Sales ($billion)')
plt.ylabel('Market Value ($billion)')
plt.title('Scatter Plot: Sales vs. Market Value')
plt.show()
```

# 3:- line chart

```python
profits_over_rank= df.groupby('Global Rank')['Profits ($billion)'].sum()

plt.figure(figsize=(10, 5))
profits_over_rank.plot(kind='line')
plt.xlabel('Global Rank')
plt.ylabel('Total Profits ($billion)')
plt.title('Total Profits Over rank')
plt.show()
```

# 4:- histograph

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Profits ($billion)'], bins=20, edgecolor='k')
plt.xlabel('Profits ($billion)')
plt.ylabel('Frequency')
plt.title('Distribution of Company Profits')
plt.show()
```

# 5:- pie graph 

```python
continent_counts = df['Continent'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(continent_counts, labels=continent_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Companies by Continent')
plt.show()
```

# 6:- boxplot

```python
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['Continent'] == 'Asia']['Sales ($billion)'], 
             df[df['Continent'] == 'Europe']['Sales ($billion)'], 
             df[df['Continent'] == 'North America']['Sales ($billion)']], 
            labels=['Asia', 'Europe', 'North America'])
plt.xlabel('Continent')
plt.ylabel('Sales ($billion)')
plt.title('Sales Distribution by Continent (Box Plot)')
plt.show()
```

# 7:- Bar graph

```python
top_10_sales_profits = df.nlargest(10, 'Sales ($billion)')[['Company', 'Sales ($billion)', 'Profits ($billion)']]

plt.figure(figsize=(12, 6))
plt.bar(top_10_sales_profits['Company'], top_10_sales_profits['Sales ($billion)'], label='Sales')
plt.bar(top_10_sales_profits['Company'], top_10_sales_profits['Profits ($billion)'], label='Profits', bottom=top_10_sales_profits['Sales ($billion)'])
plt.xlabel('Company')
plt.ylabel('Amount ($billion)')
plt.title('Sales and Profits for Top 10 Companies')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
