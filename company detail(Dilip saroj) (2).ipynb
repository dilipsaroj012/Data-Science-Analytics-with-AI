```python
from bs4 import BeautifulSoup
import pandas as pd 
import requests
from html.parser import HTMLParser
```

```python
url='https://www.sec.gov/divisions/corpfin/internatl/alpha2001.htm'
```

```python
data=requests.get(url)
data
```

```python
soup=BeautifulSoup(data.text,"html.parser")
soup.prettify
```

```python
table=soup.find('table')
table
```

```python
headers=table.find_all("b",class_="blue")
print(headers)
```

```python
titles=[]
for i in headers:
    title=i.text
    titles.append(title)
```

```python
titles
```

```python
df =pd.DataFrame(columns=titles)
df
```

```python
rows=table.find_all("tr")
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    print(data)
    row=[tr.text for tr in data]
    print(row)
    l=len(df)
    df.loc[l]=row
print(df)
    
```

```python
company=pd.DataFrame(df)
```
